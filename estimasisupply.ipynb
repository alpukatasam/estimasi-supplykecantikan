```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('supply_chain_data.csv')
```

# Deskripsi Dataset

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

# Visualisasi Data

```python
df=df.select_dtypes(include=[float,int])
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah Terjual(Number of products) Berdasarkan Ketersediaan(Availability)

```python
models = df.groupby('Availability').count()[['Number of products sold']].sort_values(by='Number of products sold',ascending=True).reset_index()
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Availability'], y=models['Number of products sold'], color='royalblue')
plt.xticks(rotation=100)
```

# Jumlah memimpin(Lead times) berdasarkan Level stok(Stock levels)

```python
engine = df.groupby('Stock levels').count()[['Lead times']].sort_values(by='Lead times').reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['Stock levels'], y=engine['Lead times'], color='royalblue')
```

# distribusi pendapatan dihasilkan (Revenue generated)

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Revenue generated'])
```

# distribusi jumlah pesanan (Order quantities)

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Order quantities'])
```

# seleksi fitur

```python
features = ['Availability','Number of products sold','Stock levels','Lead times','Revenue generated','Order quantities']
x = df[features]
y = df['Price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
##'Availability','Number of products sold','Stock levels','Lead times','Revenue generated','Order quantities'
##      55    ,           802      ,            58      ,       7    ,  8661.996792392383 , 96
input_data = np.array([[55,802,58,7,8661.996792392383,96]])

prediction = lr.predict(input_data)
print('Prediksi harga supply :', prediction)
```
